```python
import numpy as np
from sklearn.metrics import confusion_matrix

y_true = np.array(['cat','dog','dog','cat','fish','dog','fish'])
y_pred = np.array(['cat','cat','cat','cat','fish','dog','fish'])

confusion_matrix(y_true, y_pred, labels=['cat','dog','fish'])
```
